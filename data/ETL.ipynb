```python
import io
import requests
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/virtualdvid/Statistics/master/data/EGM_Lectores.csv'
file = requests.get(url).text
```

```python
df = pd.read_csv(io.StringIO(file), skiprows=4, header=1)
df.head()
```

```python
df.iloc[:,0:3] = df.iloc[:,0:3].fillna(value=None, method='ffill')
df.iloc[:,3:4] = df.iloc[:,3:4].fillna('PoblaciÃ³n')
df.head()
```

```python
df.rename(columns={df.columns[0]: "CatDemog",
                   df.columns[1]: "Demog",
                   df.columns[2]: "MediciÃ³n",
                   df.columns[3]: "Vehiculo"}, inplace = True)
df.head(0)
```

```python
df2 = pd.melt(df, id_vars=list(df.columns)[:4], value_vars=list(df.columns)[4:], var_name='Ola', value_name='Lectores')
df2.head()
```

```python
df2.iloc[:,4:5] = df2['Ola'].str.split().str.get(6) + '-' + df2['Ola'].str.split().str.get(3).str.replace('a','')
df2.iloc[:,5:6] = df2['Lectores'] * 1000
df2['Filtro'] = "Lectores"
df2.head()
```

```python
df2.to_csv('EGM2016.csv', sep=',', encoding='utf-8')
```
